```python
#Importamos pandas y nunpy
import pandas as pd #Libreria para analisis de datos
import numpy as np #Libreria para operaciones matematicas avanzadas
import matplotlib.pyplot as plt #Libreria para graficos
import seaborn as sns #Libreria para graficos estadisticos

#configuramos pandas para mostrar todas las columnas en la consola
pd.set_option('display.max_columns', None) #Configura pandas para mostrar todas las columnas del DF
pd.set_option('display.width', None) #Ancho automatico para evitar que las columnas se corten
pd.set_option('display.max_colwidth', 10) #Maximo ancho de columna.
```

1. Carga e inspeccion inicial

```python
##Importamos el dataset desde Github

###Leemos el dataset con pandas y lo almacenamos en un DF. Especificamos la codificacion en latino y el separador de punto y coma.
###Es importante hacer esto para evitar problemas de lectura, especialmente con caracteres especiales.
df = pd.read_csv("Siniestros_nacionales_aleatorio_trim6.csv", encoding='latin-1', sep=';')

print("\n Archivo cargado exitosamente !")
print(f"El dataframe se llama 'df'")

###Mostramos las primeras 5 filas del dataset.
print("\n Mostramos las primeras 5 filas del dataset:")
df.head()
```

```python
###Mostramos las ultimas 5 filas del dataset.
print("\n ULTIMAS 5 FILAS DEL DATASET:")
df.tail()
```

```python
##  #Usamos df.info() para obtener informacion general sobre el dataset, como el numero de filas, columnas, tipos de datos y valores nulos.
df.info()
```

```python
### Usamos df.describe() para obtener estadisticas descriptivas de las columnas numericas del dataset, como la media, desviacion estandar, minimos, maximos y percentiles.
df.describe() 
```

```python
###Usamos df.isnull().sum() para contar el numero de valores nulos en cada columna del dataset. Esto nos ayuda a identificar si hay datos faltantes que puedan afectar nuestro analisis.
df.isnull().sum()
```

```python
###Usamos el metodo drop() de pandas para eliminar la columna 'informacion Adicional'.
###columns=['Información Adicional'] especifica la columna a eliminar, y inplace=True indica que la eliminacion se realice en el mismo dataframe sin necesidad de asignarlo a una nueva variable.
df.drop(columns=['Información Adicional'], inplace=True)
```

```python
### Renombramos las columnas 'Año de Ocurrencia' , 'Mes de Ocurrencia' y 'Día de Ocurrencia' a 'Año_Ocurrencia', 'Mes_ocurrencia' y 'Día_Ocurrencia' respectivamente, para simplificar su uso en el analisis.
df.rename(columns={'Año de Ocurrencia': 'Año_Ocurrencia', 'Mes de Ocurrencia': 'Mes_ocurrencia', 'Día de Ocurrencia': 'Día_Ocurrencia'}, inplace=True)
```

```python
### Usamos df.isna().sum() para verificar nuevamente si hay valores nulos en el dataset despues de eliminar la columna 'Información Adicional'.
df.isna().sum()
```

```python
### Usamos df['Tiempo Inoperativo (Horas)'].describe() para obtener estadisticas descriptivas de la columna 'Tiempo Inoperativo (Horas)', como la media, desviacion estandar, minimos, maximos y percentiles.
df['Tiempo Inoperativo ( HORAS )'].describe(include='all')
```

```python
###Convertimos la columna 'Fecha Siniestro' a formato datetime para facilitar su analisis usando el metodo to_datetime() de pandas. El argumento errors='coerce' convierte los valores no convertibles a NaT (Not a Time), lo que nos permite manejar fechas invalidas sin generar errores.
df['Fecha Siniestro'] = pd.to_datetime(df['Fecha Siniestro'], errors='coerce', dayfirst=True)
```

```python
###Rellenar valores nulos en columnas 23 y 24.
df['Descripción de Causa'] = df['Descripción de Causa'].fillna('No hay descripción de causa')
df['Descripción Daños'] = df['Descripción Daños'].fillna('No hay descripción de daños')
```

```python
##Comprobacion si existen valores nulos en las columnas 23 y 24 despues de rellenar los valores nulos.
df[['Descripción de Causa' , 'Descripción Daños']].head(30) 
```

```python
##Enumerate permite iterar sobre una secuencia y obtener tanto el indice como el valor de cada elemento. En este caso, se utiliza para imprimir el indice y el nombre de cada columna del dataframe.
for i, columna in enumerate(df.columns, start=1):
    print(f"{i}. {columna}")
```

```python
###Miramos cuantas columnas son de tipo object y obtenemos una lista de las columnas de tipo object.
columnas_object = df.dtypes[df.dtypes == 'object'].index
print(f"\n El dataset tiene {len(columnas_object)} columnas de tipo object")
```

```python
###Miramos cuantas columnas numericas son de tipo int64 o float64 y extraemos sus nombres.
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(f"\nEl dataset tiene {len(numeric_cols)} columnas numéricas")
print(f"Las columnas numéricas son: {list(numeric_cols)}")
```

```python
###Identificar las columnas que no tienen ningun valor nulo.
columnas_registradas = df.columns[df.isnull().sum() == 0].tolist()
print(f"\n Las columnas sin valores nulos son: {list(columnas_registradas)}")   
```

ESTADISTICA SIMPLE

```python
###Calculamos la estadistica descriptiva del DF
Total_siniestros=len(df)
print(f"\nEl total de siniestros registrados es: {Total_siniestros}")

tiempo=df['Tiempo Inoperativo ( HORAS )']

print(f"Promedio: {tiempo.mean(): .2f} horas")
print(f"Mediana: {tiempo.median(): .2f} horas")
print(f"Valor mínimo: {tiempo.min(): .2f} horas")
print(f"Valor máximo: {tiempo.max(): .2f} horas")
print(f"Desviación estándar: {tiempo.std(): .2f} horas")
```

```python
###Realizamos algunas estadisticas personalizadas.
print("\nESTADISTICAS ADICIONALES")

print(f"Suma total de horas inoperativas: {tiempo.sum()} horas")
print(f"Rango (max -min): {tiempo.max() - tiempo.min()} horas")
```

```python
## 1. Calcular cuantos siniestros tuvieron tiempo inoperativo = 0 horas.
siniestros_cero = (df['Tiempo Inoperativo ( HORAS )'] == 0).sum()
print(f"\nTotal de siniestros con tiempo inoperativo igual a 0 horas es: {siniestros_cero}")

## 2. Calcular el porcentaje de siniestros con tiempo inoperativo mayores a 0.
promedio_tiempo_operativo = df[df['Tiempo Inoperativo ( HORAS )'] > 0]['Tiempo Inoperativo ( HORAS )'].mean()
print(f"Promedio de tiempo inoperativo (horas): {promedio_tiempo_operativo: .2f} horas")

## 3. Encuentra cuantos siniestros superan las 100 horas inoperativas.
siniestros_mas_100_horas = df[df['Tiempo Inoperativo ( HORAS )'] > 100]
print(f"Total de siniestros con tiempo inoperativo > 100 horas: {len(siniestros_mas_100_horas)}")

## 4. Calcular el percentol 90 del tiempo inoperativo (90% de los datos esta por debajo).
percetil_90 = np.percentile(df['Tiempo Inoperativo ( HORAS )'], 90)
print(f"Percentil 90 del tiempo inoperativo: {percetil_90:.2f} horas")
```

FILTROS BASICOS

```python
##Aca realizaremos los filtros donde el conductor fue el culpable.
print("\nFILTRO 1: Conductor culpable")
conductor_culpable = df[df['Culpabilidad Conductor'] == 'SI']
print(f"Total de siniestros con conductor culpable: {len(conductor_culpable)}")
print(f"Porcentaje: {(len(conductor_culpable) /len(df)) * 100:.1f}%")
print("\nPrimeros 3 registros:")
print(conductor_culpable[['Fecha Siniestro', 'Nombre Conductor','Causa Siniestro']].head(3))

##Aca realizaremos los siniestros que occurrieron en el departamento de Antioquia.
print("\nFILTRO 2: Siniestros en Antioquia")
siniestros_antioquia = df[df['DEPARTAMENTO'] == 'ANTIOQUIA'] #Corregido: 'Departaamento' a 'DEPARTAMENTO'
print(f"Total de siniestros en Antioquia: {len(siniestros_antioquia)}")
print(f"Porcentaje del total: {(len(siniestros_antioquia) / len(df)) * 100:.1f}%")

##Aca realizaremos el filtro de siniestros de vehiculos tipo 'Compactador'.
print("\nFILTRO 3: Siniestros de vehículos tipo 'Compactador'")
siniestros_compactador = df[df['Tipo de vehículo'] == 'Compactador'] #Corregido: 'Tipo Vehiculo' a 'Tipo de vehículo'
print(f"Total de siniestros de vehículos tipo 'Compactador': {len(siniestros_compactador)}")
print(f"\nPorcentaje del total: {(len(siniestros_compactador) / len(df)) * 100:.1f}%")

df['Tipo de vehículo'].unique()
```

```python
## 1. Filtrar los siniestros donde el conductor no fue el culpable.
siniestros_no_culpable=df[df['Culpabilidad Conductor'] == 'NO']
print(f"\nTotal de siniestros donde el conductor no fue el culpable: {len(siniestros_no_culpable)}")
print(f"Porcentaje del total: {(len(siniestros_no_culpable) / len(df)) * 100:.1f}%")

## 2. Filtrar siniestros en la ciudad de Ibague.
siniestros_ibague = df[df['CIUDAD'] == 'IBAGUE']
print(f"\nTotal de siniestros en la ciudad de Ibague: {len(siniestros_ibague)}")
print(f"Porcentaje del total: {(len(siniestros_ibague) / len(df )) * 100:.1f}%")

## 3. Filtrar siniestros reportados a la aseguradora (SI) que sean de MEDELLIN
siniestros_reportados_medellin = df[(df['Reportado Aseguradora'] == 'SI') & (df['CIUDAD'] == 'MEDELLIN')]
print(f"\nTotal de siniestros reportados a la aseguradora en Medellín: {len(siniestros_reportados_medellin)}")
print(f"Porcentaje del total: {(len(siniestros_reportados_medellin) / len(df)) * 100:.1f}%")    

## 4. Filtrar siniestro en turno nocturno en el departamento de atlantico con culpabilidad (SI) del conductor.
siniestros_nocturno_atlantico_culpable = df[(df['Turno'] == 'NOCTURNO') & (df['DEPARTAMENTO'] == 'ATLANTICO') & (df['Culpabilidad Conductor'] == 'SI')]
print(f"\nTotal de siniestros en turno nocturno en el departamento de Atlántico con culpabilidad del conductor: {len(siniestros_nocturno_atlantico_culpable)}")
print(f"Porcentaje del total: {(len(siniestros_nocturno_atlantico_culpable) / len(df)) * 100:.1f}%")
```
